```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df =pd.read_csv('canada_per_capita_income.csv')
print(df.describe())
print(df.isna().sum())                            #checking for null values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X=df.drop(columns=['per capita income (US$)'])
Y=df['per capita income (US$)']
model=LinearRegression()
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.15)
model.fit(X_train,Y_train)
```

```python
ans=model.predict([[2020]])
print(ans)
```

```python
print(model.score(X,Y))
ans1=model.predict(X_test)
```

```python
import matplotlib as plt
import seaborn as sns

result=pd.DataFrame(columns=['YEAR','INCOME'])
result['YEAR']=X_test['year']
result['INCOME'] =ans1
sns.lineplot(x='YEAR',y='INCOME',data=result)
```
